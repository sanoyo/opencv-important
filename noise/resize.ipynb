```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

### リサイズ

```python
img = cv2.imread("../images/sample5.jpg")
size = img.shape

# 画像を2分の1にする
my_img = img[: size[0] // 2, : size[1] // 2]
my_img = cv2.resize(my_img, (my_img.shape[1] * 2, my_img.shape[0] * 2))

plt.imshow(my_img)
```

### 回転・反転

```python
img = cv2.imread("../images/sample5.jpg")

# 画像を反転させる
# 0 → x軸、正→y軸、負→両方
my_img = cv2.flip(img, 0)

plt.imshow(my_img)
```

### アフィン変換

```python
img = cv2.imread("../images/sample6.jpg")

# warpAffine()を用いるのに必要な行列をつくる
# 第一引数が回転の中心、第二引数は回転角度、第三引数は倍率
mat = cv2.getRotationMatrix2D(tuple(np.array([img.shape[1], img.shape[0]]) / 2), 90, 2.0)

# 第一引数が変換したい画像、第二引数が上で生成した行列(mat)、第三引数サイズ
my_img = cv2.warpAffine(img, mat, img.shape[:2])

plt.imshow(my_img)
```
